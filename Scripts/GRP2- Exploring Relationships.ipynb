### a. Correlation

```python
# Create a correlation matrix using pandas

apt.corr()
```

#### Create a correlation heatmap using Seaborn

```python
apt.columns
```

```python
# Create a subset excluding categorical variables

sub = apt[['utility cost', 'picturecount', 'pricetrend', 'year constructed', 'base rent', 'number of rooms']]
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), annot = True, ax = ax) 
```

#### Note: There is a strong potitive correlation between utility cost and base rent, meaning the higher the base rent the higher the utility cost.   There is a moderately-high positive correlation between number of rooms and utility cost suggesting that more rooms means a higher utility cost.  Other moderate correlations exist between price trend and base rent, suggesting that the higher the base rent the higher the price trend, and number of rooms and base rent, implying that the higher the base rent the more rooms in the unit.  There are no other significant relationships.

### b. Scatterplot

```python
# Create a scatterplot for the "pricetrend" and "base rent" columns in seaborn

sns.lmplot(x = 'pricetrend', y = 'base rent', data = apt)
```

#### Note: Negative trends show rental lower rates.  As the trend increases, we see a wider range of rates with the a wide range of rates centeresed aroung 5%.  Higher rental rates tail off as price trend passes 10%.  Highest rent values are centered around a price trend of 5-10%.

```python
# Create a scatterplot for the "pricetrend" and "base rent" columns in seaborn

sns.lmplot(x = 'base rent', y = 'pricetrend', data = apt)
```

```python
# Create a scatterplot for the "number of rooms" and "base rent" columns in seaborn

sns.lmplot(x = 'base rent', y = 'number of rooms', data = apt)
```

#### Note:  There is a distinct trend line that demonstrates that as rental rates increase, the number of rooms increases.

### c. Pair Plots

```python
# Create a pair plot 

g = sns.pairplot(sub)
```

#### Note: Base rent has two peaks in terms of years.  Most expensive rents and utilities are centered around post-2000 and 1900.  Rentals built more recently have a wider range (+/-) of pricetrend than older (pre-1850) rentals, whose pricetrend is centered more aroun 0-8%.  We also see that base rent increases toward 5 rooms. We also see that more recently built rentals tend to have more rooms.
